## Orbital solutions
Calculate orbital cycles using the Python package "ClimLab" (this uses solutions originally published by Berger 1991

See https://climlab.readthedocs.io/en/latest/api/climlab.solar.orbital_cycles.html

```python
from climlab.solar.orbital import OrbitalTable
from climlab.solar.insolation import daily_insolation
years = np.linspace(-5000, 0, 5001)

# subset of orbital parameters for specified time
orb = OrbitalTable.interp(kyear=years).to_dataframe()
orb['kyear']=orb.index
orb['kyear']=orb['kyear']*-1
```
